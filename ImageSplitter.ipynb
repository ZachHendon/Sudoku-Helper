```python
import numpy
from skimage import io
from matplotlib import pyplot as plt
```

```python
square = io.imread('sudokuBoardnumBoxes.jpg')
```

```python
m, n = square.shape[:numBoxes]
```

```python
io.imshow(square)
```

```python
horizontalLines = []
verticalLines = []

for i in range(m):
    pixelCount = 0

    for j in range(n):
        if square[i, j][0] < 250:
            pixelCount += 1

    if pixelCount > square.shape[1] / 3:
        horizontalLines.append(i)

for j in range(n):
    pixelCount = 0

    for i in range(m):
        if square[i, j][0] < 250:
            pixelCount += 1

    if pixelCount > square.shape[0] / 3:
        verticalLines.append(j)

print(horizontalLines)
print(verticalLines)
```

```python
innerHorizontalLines = []
innerVerticalLines = []


for i in range(len(horizontalLines)-1):
    if horizontalLines[i] != horizontalLines[i+1]-1:
        innerHorizontalLines.append(horizontalLines[i])
        innerHorizontalLines.append(horizontalLines[i+1])
print(innerHorizontalLines)

for j in range(len(verticalLines)-1):
    if verticalLines[j] != verticalLines[j+1]-1:
        innerVerticalLines.append(verticalLines[j])
        innerVerticalLines.append(verticalLines[j+1])
print(innerVerticalLines)
```

```python
numBoxes = int((len(innerHorizontalLines) / 2) ** 2)
print(numBoxes)
print(len(innerVerticalLines))
```

```python
squareGrid = square.copy()

for hLine in innerHorizontalLines:
    squareGrid[hLine] = 0
    

for vLine in innerVerticalLines:
    squareGrid[:, vLine] = 0

io.imshow(squareGrid)
```
